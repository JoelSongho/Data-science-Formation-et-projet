```python
import pandas as pd

# Chargement du fichier
df = pd.read_csv("sample_data.csv")

# Aperçu des premières lignes
df.head(10)
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Remplir les salaires manquants par la moyenne
df['sales'] = df['sales'].fillna(df['sales'].mean())

# Supprimer les lignes où le genre est manquant
df = df.dropna(subset=['category'])
```

```python
df = df.drop_duplicates()
df
```

```python
df.groupby('region')['sales'].mean()
```

```python
df['age'].hist(bins=10)
```

```python
# Exemple : normaliser les salaires
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Salaire_norm'] = scaler.fit_transform(df[['sales']])
```

```python
# Ici je supprime tout les ligne vide de la colone
df= df.dropna(subset=['age'])

#Ici je change le type de la colonne age de float en int
df['age']=df['age'].astype(int)
df.info()
```
